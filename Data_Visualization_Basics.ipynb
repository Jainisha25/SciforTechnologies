```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Sample data
data = np.random.rand(100, 2)
```

```python
# Bar Chart with Color and Width
plt.bar(['A', 'B', 'C'], [3, 7, 2], color=['blue', 'green', 'orange'], width=0.5)
plt.title('Bar Chart with Color and Width')
plt.show()
```

```python
# Pie Chart with Custom Colors
colors = ['gold', 'lightcoral', 'lightskyblue']
plt.pie([30, 50, 20], labels=['Category 1', 'Category 2', 'Category 3'], autopct='%1.1f%%', colors=colors)
plt.title('Pie Chart')
plt.show()
```

```python
# Histogram with Color, Edgecolor, and Transparency
plt.hist(data[:, 0], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.show()
```

```python
# Sample categorical data
categories = ['Category 1', 'Category 2', 'Category 3']
data = np.random.rand(len(categories), len(categories))

# Custom color map for categories
colors = sns.color_palette("husl", len(categories))

# Simple Categorical Heatmap
sns.heatmap(data, cmap=colors, annot=True, fmt=".2f", xticklabels=categories, yticklabels=categories)
plt.title('Categorical Heatmap with Custom Colors')
plt.show()
```

```python
# Scatterplot with Color, Size, and Transparency
plt.scatter(data[:, 0], data[:, 1], c='purple', s=30, alpha=0.5)
plt.title('Scatterplot')
plt.show()
```

```python
# Box Plot with Color and Notch
sns.boxplot(x=data[:, 0], color='lightgreen', notch=True)
plt.title('Box Plot')
plt.show()
```

##Iris Dataset:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python

# Bar Graph (Sepal Length by Species)
plt.figure(figsize=(8, 6))
sns.barplot(x="species", y="sepal_length", data=iris, ci=None)
plt.title("Bar Graph of Sepal Length by Species")
plt.show()
```

```python
# Pie Chart (Distribution of Species)
plt.figure(figsize=(8, 6))
iris['species'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Pie Chart of Species Distribution")
plt.ylabel("")  # Remove the default 'species' label
plt.show()
```

```python

# Histogram (Distribution of Sepal Length)
plt.figure(figsize=(8, 6))
sns.histplot(iris['sepal_length'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
# Heatmap (Correlation Matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
# Scatterplot (Sepal Length vs Sepal Width)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris, palette="viridis")
plt.title("Scatterplot of Sepal Length vs Sepal Width")
plt.show()
```

```python
# Boxplot (Petal Length by Species)
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="petal_length", data=iris, palette="pastel")
plt.title("Boxplot of Petal Length by Species")
plt.show()
```
